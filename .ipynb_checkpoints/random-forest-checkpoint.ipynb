# Gold Price Prediction (GLD) â€“ Random Forest Regressor

### **Project Objective**  
#### Build a machine learning model to predict the gold price (column `GLD`) using related variables (SPX, USO, SLV, EUR/USD)

## Libraries & Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## 1. Data Loading and Initial Inspection

```python
df = pd.read_csv('gld_price_data.csv')

df.head()
```

```python
df = df.drop('Date', axis=1)

print("Dataset dimensions (rows, columns):", df.shape)
print("\nData types:\n", df.dtypes)
print("\nMissing values per column:\n", df.isna().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())
print("\nDescriptive statistics:\n", df.describe().round(2))
```

## 2. Exploratory Data Analysis (EDA)

```python
corr = df.corr()
plt.figure(figsize=(8, 8))
sns.heatmap(corr, cbar=True, square=True, fmt='.1f', annot=True, cmap='Blues')
```

```python
# Correlation values of gold
print(corr['GLD'])
```

```python
# Distribuition of gold price

sns.displot(df['GLD'], kind='hist', kde=True, color='blue')
```

## 3. Data Preprocessing and Feature Engineering

```python
X = df.drop(['GLD'], axis=1)
y = df['GLD']
```

## 4. Data Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## 5. Model Training

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)
```

## 6. Model Evaluation

```python
y_pred = rf.predict(X_test)
print(y_pred)
```

```python
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"RÂ² score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

## 7. Final Evaluation on Test Set

### - **RÂ²**: 0.9901 -> The model explains **99.01%** of the variance in gold prices (GLD).

### - **RMSE**: Root Mean Squared Error â‰ˆ **2.31 dollars** per prediction. 

### - **MAE**: Mean Absolute Error â‰ˆ **1.35 dollars** per prediction.
